```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans, MiniBatchKMeans
from sklearn.decomposition import PCA
from sklearn import datasets

iris = datasets.load_iris()
X = iris.data
y = iris.target

scaler = StandardScaler()
X_std = scaler.fit_transform(X)
```

```python
kmeans_cluster = KMeans(n_clusters=3, random_state=123)

%timeit kmeans_cluster.fit(X_std)
y_pred = kmeans_cluster.fit_predict(X_std)
```

```python
pca = PCA(n_components=2).fit_transform(X_std)

plt.figure(figsize=(10,5))
colours = 'rbg'
for i in range(pca.shape[0]):
    plt.text(pca[i, 0], pca[i, 1], str(y_pred[i]),
             color=colours[y[i]],
             fontdict={'weight': 'bold', 'size': 50}
        )

plt.xticks([])
plt.yticks([])
plt.axis('off')
plt.show()
```

```python
kmeans_cluster = KMeans(n_clusters=4, random_state=123)

%timeit kmeans_cluster.fit(X_std)
y_pred = kmeans_cluster.fit_predict(X_std)
```

```python
pca = PCA(n_components=2).fit_transform(X_std)

plt.figure(figsize=(10,5))
colours = 'rbg'
for i in range(pca.shape[0]):
    plt.text(pca[i, 0], pca[i, 1], str(y_pred[i]),
             color=colours[y[i]],
             fontdict={'weight': 'bold', 'size': 50}
        )

plt.xticks([])
plt.yticks([])
plt.axis('off')
plt.show()
```

```python
kmeans_cluster = KMeans(n_clusters=2, random_state=123)

%timeit kmeans_cluster.fit(X_std)
y_pred = kmeans_cluster.fit_predict(X_std)
```

```python
pca = PCA(n_components=2).fit_transform(X_std)

plt.figure(figsize=(10,5))
colours = 'rbg'
for i in range(pca.shape[0]):
    plt.text(pca[i, 0], pca[i, 1], str(y_pred[i]),
             color=colours[y[i]],
             fontdict={'weight': 'bold', 'size': 50}
        )

plt.xticks([])
plt.yticks([])
plt.axis('off')
plt.show()
```

When value of k is equal to 4, we can see that K-means seem to succesfully identify the linearly separable two classes but it is not that succesfull for the red class. I think that the best value of k is 3.
